### 資料清理 - 輸出所有類別新聞 allnews.csv

```python
###
```

```python
import pandas as pd
import os
```

```python
data_path='../dataset_include_content/regular_dataset/'
```

```python
listOfFiles=os.listdir(data_path)
flist=[f for f in listOfFiles if f.endswith("json")]
for file in flist:
    print(file)
```

```python
topic_dic={'旅遊':1,
       '蘋果專區':2,
       '美食':3,
       'AI與大數據':4,
       '職涯':5,
       '美妝保養':6,
       '理財':7,
       '房地產':8,
       '汽車':9,
       '醫療健康':10,
      '財經':11,
      '區塊鏈技術':12,
          '3C產品':13,
          '無人駕駛':14}
```

```python
topic=[]
topic_name=[]
title=[]
content=[]
cbodyurl=[]
url=[]
txtlen=[]
for f in flist:
    df=pd.read_json(os.path.join(data_path,f))
    
    for i in range(len(df)):
        topic_number=topic_dic.get(f.split('_')[0])
        topic_name_txt=f.split('_')[0]
        title_txt=df['title'][i]
        content_txt=df['content'][i]
        url_txt=df['url'][i]
        cbodyurl_txt=df['cbodyurl'][i]
        txtlen_number=len(df['content'][i])
        
        topic.append(topic_number)
        topic_name.append(topic_name_txt)
        title.append(title_txt)
        content.append(content_txt)
        url.append(url_txt)
        cbodyurl.append(cbodyurl_txt)
        txtlen.append(txtlen_number)
    
    
outdf=pd.DataFrame({
    'topic':topic,
    'topic_name':topic_name,
    'title':title,
    'content':content,
    'txtlen':txtlen,
    'url':url,
    'cbodyurl':cbodyurl
    })
    
outdf.info()
#outdf.to_csv('allnews.csv',index=False)
```

```python
outdf=pd.DataFrame({'topic':outdf['topic'],'topic_name':outdf['topic_name'],
                    'title':outdf['title'],'content':outdf['content'],'txtlen':outdf['txtlen']})
outdf.head()
```

### 輸出各類別的Train/Test資料

```python
trin=[]
test=[]
for key in topic_dic.keys():
    ttdf=outdf[outdf.topic==topic_dic.get(key)]
    seperate_index=int(len(ttdf)*0.9)
    trin.append(ttdf.iloc[:seperate_index])
    test.append(ttdf.iloc[seperate_index:])
    
traindf=pd.concat(trin)
testdf=pd.concat(test)

traindf.info()
testdf.info()
```
